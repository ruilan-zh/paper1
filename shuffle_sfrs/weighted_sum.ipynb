```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
b = np.arange(1, 100,0.001)
```

```python
mu = 100
sigma = 1
```

```python
S_mu = np.sum(b*mu)/(len(b)*mu)
print(S_mu)
S_list = []
for i in range(100):
    L = np.random.normal(loc=mu, scale=sigma, size=len(b))
    S = np.sum(b*L)/np.sum(L)
    S_list.append(S)
```

```python
plt.hist(S_list)
plt.vlines(S_mu, ymin=0, ymax=20, color="red")
#plt.xlim(50.495, 50.505)
```

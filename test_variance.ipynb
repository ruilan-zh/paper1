```python
data_cent = np.loadtxt(f"{odir}/sfr-halomass_central.txt")
```

```python
mvir_cent = 10**data_cent[:,1]
sfr_cent = 10**data_cent[:,5]
```

```python
sfr_cent_squared = sfr_cent**2
sfr_cent_sum = np.sum(sfr_cent)
shot_cent = np.sum(sfr_cent_squared)/sfr_cent_sum**2
print(shot_cent)
```

```python
lum_cent = sfr_cent * 2 * 10**41
```

```python
lum_cent_squared = lum_cent**2
lum_cent_sum = np.sum(lum_cent)
shot_cent_lum = np.sum(lum_cent_squared)/lum_cent_sum**2
print(shot_cent_lum)
```

```python
sfr_group_squared = sfr_group**2
sfr_group_sum = np.sum(sfr_group)
shot_group = np.sum(sfr_group_squared)/sfr_group_sum**2
print(shot_group)
```

```python
ratio_shot = shot_cent/shot_group
print(ratio_shot)
```

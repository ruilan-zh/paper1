```python
import seaborn as sns
palette_tab10 = sns.color_palette("colorblind", 10)
color_list = palette_tab10[1:2] + palette_tab10[2:3] + palette_tab10[4:5]
```

```python
figsize=(2*10/3, 2*2.5)
fig1 = plt.figure(figsize=figsize)
frame1 = fig1.add_axes((.1,.3,.8,.6))
plt.plot(z_list, np.log10(tot_sfrd), label="Total", color=color_list[0])
plt.plot(z_list, np.log10(cent_sfrd), label="Central", color=color_list[1])
plt.plot(z_list, np.log10(sat_sfrd), label="Satellite", color=color_list[2])
plt.ylabel("SFRD [M$_\odot$yr$^{-1}$ h$^{3}$cMpc$^{-3}$]")
plt.legend()

# plt.plot(z_list, np.log10(tot_sfrd1), label="Total", color=color_list[0], linestyle="dashed")
# plt.plot(z_list, np.log10(cent_sfrd1), label="Central", color=color_list[1], linestyle="dashed")
# plt.plot(z_list, np.log10(sat_sfrd1), label="Satellite", color=color_list[2], linestyle="dashed")

linestyle2 = "dashed"
plt.plot(z_list2, np.log10(tot_sfrd2), label="Total", color=color_list[0], linestyle=linestyle2)
plt.plot(z_list2, np.log10(cent_sfrd2), label="Central", color=color_list[1], linestyle=linestyle2)
plt.plot(z_list2, np.log10(sat_sfrd2), label="Satellite", color=color_list[2], linestyle=linestyle2)


#vline = plt.vlines(4.4, ymin=-4, ymax=-0.5, linestyle="dotted", color="grey", label="100 DM particles")
plt.ylim(-4,-0.2)
frame2 = fig1.add_axes((.1,.1,.8,.2), sharex=frame1) 
plt.plot(z_list, 100*sat_sfrd/tot_sfrd, color=color_list[2])
#plt.plot(z_list, 100*sat_sfrd1/tot_sfrd1, color=color_list[2], linestyle="dashed")

plt.plot(z_list2, 100*sat_sfrd2/tot_sfrd2, color=color_list[2], linestyle=linestyle2)

from matplotlib.lines import Line2D
#dashed_line = Line2D([],[],color="grey", linestyle="dashed", label="SubhaloFlag=1")

solid_line = Line2D([],[],color="grey", linestyle="solid", label="TNG300-1")
dashed_line = Line2D([],[],color="grey", linestyle="dashed", label="TNG50-1")
dotted_line = Line2D([],[],color="grey", linestyle="dotted", label="TNG50-1")
ax = plt.gca()
leg1 = plt.legend(handles=[solid_line, dashed_line], bbox_to_anchor=(0.44,0.9))
ax.add_artist(leg1)

#vline = plt.vlines(4.4, ymin=6, ymax=36, linestyle="dotted", color="grey", label="100 DMparticles")
#leg2 = plt.legend(handles=[vline], loc="lower right",bbox_to_anchor=(1,0.2))
#ax.add_artist(leg2)

plt.ylabel("% Satellite")
plt.xlabel("Redshift")
#plt.savefig("figures/SFRD_eagle_5%.pdf", format="pdf", bbox_inches="tight")
```

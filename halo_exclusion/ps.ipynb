```python
def pk_c(k, exp_param):
    pk = k**exp_param
    
    return pk
    
```

```python
def total_pk(k, exp_param, R_param):
    
    pk_tot = pk_c(k, exp_param) - result_k(k,R_parameter) - np.convolve(pk_c(k, exp_param), result_k(k,R_parameter), "same")
    return pk_tot
```

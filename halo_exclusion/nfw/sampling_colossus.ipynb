```python
from colossus.cosmology import cosmology
from colossus.halo import profile_nfw
import numpy as np
import matplotlib.pyplot as plt
```

```python
cosmology.setCosmology('planck18')
p_nfw = profile_nfw.NFWProfile(M = 1E12, c = 10.0, z = 0.0, mdef = 'vir')
```

```python
r = np.arange(0,1000,0.1)
cum_pdf = p_nfw.cumulativePdf(r, Rmax=1000)
```

```python
plt.plot(r, cum_pdf)
```

```python
profile_nfw.radiusFromPdf(M=1e13, c=4, z=1.5, mdef="vir", cumulativePdf=cum_pdf, interpolate=True, min_interpolate_pdf=0.01)
```
